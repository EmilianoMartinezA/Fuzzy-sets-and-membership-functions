```python
# a) Función de membresia para cuantificar el conjunto de la altura media
def membership_function(x, a, b, c, d):
    if x < a or x > d:
        return 0
    elif a <= x < b:
        return (x - a) / (b - a) #Pendiente positiva
    elif b <= x <= c:
        return 1
    elif c < x <= d: 
        return (d - x) / (d - c) #Pendiente negativa
    
#Definir los valores de los puntos de inflexión
a = 155
b = 165
c = 175
d = 185


#Generar los valores de x y y
x = np.linspace(140, 190, 1000)
y = [membership_function(xi, a, b, c, d) for xi in x]
```

```python

# d) Funcion de membresia para el conjunto de "cerca de 10"
def triangular_membership(x, a, b, c):
    if x < a or x > c:
        return 0 #Fuera del rango
    elif a <= x < b:
        return (x - a) / (b - a) #Pendiente positiva
    elif b <= x <= c:
        return (c - x) / (c - b) #Pendiente negativa

#Definir los valores de los puntos de inflexión
a = 8
b = 10
c = 12

#Generar la función de membresía
x = np.linspace(5, 15, 1000)
y = [triangular_membership(xi, a, b, c) for xi in x]

```

#### Función de membresia Bronceado de Color

Para el color de la piel, podemos utilizar una función de membresía trapezoidal, donde las personas más bronceadas están en el extremo derecho del espectro y las menos bronceadas en el extremo izquierdo. Supongamos que el rango de colores va de 0 (blanco) a 1 (negro), y que un tono de 0.5 es considerado "bronceado". La función de membresía trapezoidal para este conjunto difuso sería

<img src = "Captura de pantalla 2024-04-10 a la(s) 21.41.05.png">


```python


# Definir las funciones de membresía para estatura media y bronceado
def media(x):
    if x <= 150 or x >= 180:
        return 0
    elif 150 < x <= 165:
        return (x - 150) / 15
    elif 165 < x < 180:
        return (180 - x) / 15
    else:
        return 0

def bronceado(x):
    if x <= 0 or x >= 1:
        return 0
    elif 0 < x <= 0.25:
        return (x - 0) / 0.25
    elif 0.25 < x < 0.5:
        return 1
    elif 0.5 <= x <= 0.75:
        return 1
    elif 0.75 < x < 1:
        return (1 - x) / 0.25
    else:
        return 0

# Crear un conjunto de valores para x
x = np.linspace(140, 190, 1000)

# Calcular las funciones de membresía para estatura media y bronceado
y_media = [media(i) for i in x]
y_bronceado = [bronceado(i) for i in np.linspace(0, 1, 1000)]

# Calcular la función de membresía para estatura media y bronceado (usando mínimo)
y_estatura_bronceado = np.minimum(y_media, y_bronceado)

# Graficar las funciones de membresía
plt.figure(figsize=(10, 6))
plt.plot(x, y_media, label='Estatura Media')
plt.plot(np.linspace(0, 1, 1000), y_bronceado, label='Bronceado')
plt.plot(x, y_estatura_bronceado, label='Estatura Media y Bronceado')
plt.xlabel('Valor')
plt.ylabel('Grado de membresía')
plt.title('Funciones de membresía')
plt.legend()
plt.grid(True)
plt.show()
```

(b)  Draw a membership function (and hence define a fuzzy set) that quantifies the set of all short people who are “white” in color (i.e., short and white people).

Se define una funcion de membresia triangular para la parte de definir con logica fuzzy la estatura de una persona, en este caso que tan pequeña es, en un rango que va de 120cm a 150cm

<img src = "Captura de pantalla 2024-04-10 a la(s) 21.45.19.png">

Para establecer la funcion de membresia del tono de color de una persona, la funcion debera ser trapezoidal

<img src = "Captura de pantalla 2024-04-10 a la(s) 21.52.04.png">


```python

# Membership function for short height
def short_height(x):
    if x <= 120 or x >= 150:
        return 0
    elif 120 < x <= 135:
        return (x - 120) / 15
    elif 135 < x < 150:
        return (150 - x) / 15
    else:
        return 0

# Membership function for white color
def white_color(x):
    if x <= 0 or x >= 1:
        return 0
    elif 0 <= x <= 0.2:
        return (x - 0) / 0.2
    elif 0.2 < x < 0.4:
        return 1
    elif 0.4 <= x <= 1:
        return (1 - x) / 0.2
    else:
        return 0

# Create a range of values for height and color
x_height = np.linspace(110, 160, 1000)
x_color = np.linspace(0, 1, 1000)

# Calculate the membership functions for height and color
y_short_height = [short_height(i) for i in x_height]
y_white_color = [white_color(i) for i in x_color]

# Combine the membership functions using the minimum operator
y_short_white = np.minimum(y_short_height, y_white_color)

# Plot the membership function for short and white
plt.figure(figsize=(10, 6))
plt.plot(x_height, y_short_height, label='Short Height')
plt.plot(x_color, y_white_color, label='White Color')
plt.plot(x_height, y_short_white, label='Short and White')
plt.xlabel('Value')
plt.ylabel('Membership')
plt.title('Membership Function for Short and White')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar la función de membresía para color blanco
plt.figure(figsize=(10, 6))
plt.plot(x_color, y_white_color, label='Color Blanco')
plt.xlabel('Color')
plt.ylabel('Grado de membresía')
plt.title('Función de membresía para personas blancas')
plt.legend()
plt.grid(True)
plt.show()
```

(c)  Draw a membership function (and hence define a fuzzy set) that quantifies the set of all tall people who are “black” in color (i.e., tall and black people).

Se define la altura como un conjunto difuso, impliementando una funcion de membresia triangular.

<img src= "Captura de pantalla 2024-04-10 a la(s) 21.57.43.png">

Mientras que para definir el color de la piel como un conjunto difuso, se define con una funcion trapezoidal donde las personas de tono de piel mas oscura es totalmente perteneciente al conjunto.

<img src= "Captura de pantalla 2024-04-10 a la(s) 22.01.32.png">

```python
# Función de membresía para altura alta
def altura_alta(x):
    if x <= 170 or x >= 200:
        return 0
    elif 170 < x <= 185:
        return (x - 170) / 15
    elif 185 < x < 200:
        return (200 - x) / 15
    else:
        return 0

# Función de membresía para tono negro
def color_negro(x):
    if x <= 0 or x >= 1:
        return 0
    elif 0 <= x <= 0.2:
        return (x - 0) / 0.2
    elif 0.2 < x < 0.4:
        return 1
    elif 0.4 <= x <= 0.8:
        return 1
    elif 0.8 < x <= 1:
        return (1 - x) / 0.2
    else:
        return 0

# Crea un conjunto de valores para x (altura y color)
x_altura = np.linspace(160, 210, 1000)
x_color = np.linspace(0, 1, 1000)

# Calcula las funciones de membresía para altura alta y tono negro
y_altura_alta = [altura_alta(i) for i in x_altura]
y_color_negro = [color_negro(i) for i in x_color]

# Grafica la función de membresía para altura alta
plt.figure(figsize=(10, 6))
plt.plot(x_altura, y_altura_alta, label='Altura Alta')
plt.xlabel('Altura')
plt.ylabel('Grado de membresía')
plt.title('Función de membresía para Altura Alta')
plt.legend()
plt.grid(True)
plt.show()

# Grafica la función de membresía para tono negro
plt.figure(figsize=(10, 6))
plt.plot(x_color, y_color_negro, label='Color Negro')
plt.xlabel('Color')
plt.ylabel('Grado de membresía')
plt.title('Función de membresía para Color Negro')
plt.legend()
plt.grid(True)
plt.show()
```
